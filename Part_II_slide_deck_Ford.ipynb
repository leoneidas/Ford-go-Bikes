## Dataset Overview

> This data set includes information about individual rides made on the Ford-Go-Bike bike-sharing system covering the greater San Francisco Bay area for month of February 2019.This dataset includes 519,700 trips with 15 features  entailing the users from their Birth Year, User Type, Duration of Ride, Start station, Start Location and Gender among may others.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
#Load the data set and display
df=pd.read_csv('201902-fordgobike-tripdata.csv')
```

```python
#Changing data types
df['start_time']=pd.to_datetime(df['start_time'])
df['end_time']=pd.to_datetime(df['end_time'])

#Changing Duration to minutes
df['duration_minute'] = df['duration_sec']/60    
df['duration_minute'] = df['duration_minute'].astype(int)

#creat a start day column
df['start_day'] = df['start_time'].dt.day_name()

# creat age column
df['age'] = 2019 - df['member_birth_year']
df['age'] = df['age'].fillna(0).astype(int)

#Droping Null values
df.dropna(axis=0,inplace=True)
```

## The Start Day

> The visualization below shows the day the ride was made. This helps to how the comparison for the demand of the servive  through out each day of the week.



```python
#Start Day Bar Chart
plt.figure(figsize=[14, 7]);
base_color = sb.color_palette()[0]
order = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
sb.countplot(data=df, x='start_day', color=base_color, order= order)
plt.xlabel('Start Day', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.title('Bar Chart Showing The Start Day',fontsize=16);

```

>This shows that Thursday was the most common day to hire a bike with weekends having the least. This maybe due to the fact that the weekdays are work days and thus the users used the service to beat triffic or get somewhere fater.

## The Type Of Users

> This is a pie chart showing the Type of Uses if they are Subscribers or walk in Customers

```python
sorted_counts = df['user_type'].value_counts()
plt.figure(figsize=[10,7])
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90,autopct='%.2f', counterclock = False);
plt.axis('square')
plt.legend(loc='upper right');
plt.title('Pie Chart Showing Type of Users',fontsize=16);
```

>This shows that majority of the users were Subscribed to Ford-Go-Bikes constituting of 90.53%

## The Users Age Distribution
>This is a histogram showing the age distribution of the users.


```python
#The users Age
binsize = 5
bins = np.arange(0, df['age'].max()+binsize, binsize)
plt.figure(figsize=[10, 5])
plt.hist(data = df, x = 'age' ,bins=bins)
plt.xlabel('Age', size='14')
plt.ylabel('Count ', size='14')
plt.axis([-10, 140, 0, 70000])
plt.title('The Users Age Distribution', size='14');
```

>It can be seen that  the Average age of most users was between 20-45years.

## The Ride Duration
>This is ahistogram showing the ride duration

```python
binsize = 5
bins = np.arange(0, df['duration_minute'].max()+binsize, binsize)
plt.figure(figsize=[10, 5])
plt.hist(data = df, x = 'duration_minute' ,bins=bins)
plt.xlabel('Duration (min)', size='14')
plt.ylabel('Count ', size='14')
plt.axis([-10, 140, 0, 70000])
plt.title('The Ride Duration', size='14');
```

 Most rides durations were averagly between  0-25minutes
